```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context("talk")
plt.rcParams['font.family'] = "Ubuntu"
```

```python
df = pd.read_excel('./Files_II/Facebook Insights Data Export - Volt Danmark - 2020-08-03.xls', sheet_name='Key Metrics', skiprows=[1], index_col='Data')
df = df.fillna(value=0)
df.index = pd.to_datetime(df.index)
df.color = '#582c83'
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

df['Lifetime Total Likes'].plot(ax=ax, color='#582c83')

ax.set_title("Total number of page likes", loc='left')
ax.set_ylabel('Likes')
ax.set_xlabel('')

plt.tight_layout()
plt.savefig('./Figures/totalpagelikes3.png', dpi=300)

plt.show()
```

---

```python
fig, ax = plt.subplots(figsize=(8,8), nrows=3, sharex=True)

df['Daily Page Engaged Users'].plot(ax=ax[0], color='#582c83')

ax[0].set_title("Engaged users - click or story", loc='left')
ax[0].set_ylabel('Users')

df['Daily Paid Reach'].plot(ax=ax[1], color='#582c83')
ax[1].set_title('Daily paid reach', loc='left')

df['Daily Total Impressions'].plot(ax=ax[2], color='#582c83')
ax[2].set_title('Daily Total impressions', loc='left')


ax[2].set_xlabel('')

for axe in ax:
    axe.set_ylim(bottom=0)

plt.tight_layout()
plt.subplots_adjust(hspace=0.24)
plt.savefig('./Figures/engagement3.png', dpi=300)

plt.show()
```
